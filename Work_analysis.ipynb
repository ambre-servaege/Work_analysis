```python
mu = df['HREC'].mean()  # mean of distribution
sigma = df['HREC'].std()  # standard deviation of distribution

plt.figure()

# the histogram of the data
n, bins, patches = plt.hist(df['HREC'], 20, density=True)

# Formule pour la densitÃ©
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

plt.plot(bins, y, '--')
plt.xlabel('Smarts')
plt.ylabel('Probability density')
plt.title(r'Histogram : $\mu={}$, $\sigma={}$'.format(np.round(mu,2), np.round(sigma,2)))
plt.show()
```
